```python
import pandas as pd
import numpy as np
```

```python
Sales = pd.read_csv('Sales.csv')
```

```python
# fill missing value in taxAmt,Freight,TotalDue by 0
Sales[['TaxAmt','Freight','TotalDue']] = Sales[['TaxAmt','Freight','TotalDue']].fillna(0)
```

```python
Sales['SalesPersonID'] = Sales['SalesPersonID'].astype('object')
Sales['SalesPersonID'] = Sales['SalesPersonID'].fillna('Unknown')
```

```python
Sales.isna().sum().sum()
```

```python
# convert oder date ,due,ship to type date
Sales['OrderDate'] = pd.to_datetime(Sales['OrderDate'])
Sales['DueDate'] = pd.to_datetime(Sales['DueDate'])
Sales['ShipDate'] = pd.to_datetime(Sales['ShipDate'])
```

Feature Engineering

```python
# Calculate Delevery Date
Sales['DeleveryDate']= (Sales['ShipDate']-Sales['OrderDate']).dt.days
# Caluclate DiscountAmount
Sales['DiscountAmount'] = Sales['UnitPrice']*Sales['OrderQty']*Sales['UnitPriceDiscount']
# Caluclate profit
Sales['Profit'] = Sales['UnitPrice']*Sales['OrderQty']-Sales['DiscountAmount']
# Calucalte Year
Sales['Year'] = Sales['OrderDate'].dt.year
# Calculate Month
Sales['Month'] = Sales['OrderDate'].dt.month
```
